### Используемые инструменты:
1. `numpy ` — базовые вычисления и обработка пропусков;
2. `pandas` — табличная обработка данных, сводные описательные статистики;

#### Количественная оценка выбросов в числовых данных

Анализ выполняется двумя методами Z-score и IQR. Дополнительно проверяется наличие отрицательных значений в показателях, где это не имеет физического смысла (GDP, CO₂, Energy).

```python
import numpy as np
import pandas as pd
from scipy import stats

cols_check = ["co2", "co2_per_capita", "gdp", "primary_energy_consumption"]
num_cols = [c for c in cols_check if c in df_focus.columns]


def basic_summary(df, cols):
    return df[cols].apply(pd.to_numeric, errors="coerce").describe().T


def count_negatives(df, cols):
    cols = [c for c in cols if c in df.columns]
    s = df[cols].apply(pd.to_numeric, errors="coerce")
    return (s.lt(0).sum()).rename("neg_count")


def outlier_share_z(series, thr=3.0):
    s = pd.to_numeric(series, errors="coerce").dropna()
    if len(s) == 0 or s.std(ddof=0) == 0:
        return np.nan
    z = np.abs(stats.zscore(s))
    return float((z > thr).mean()) * 100


def outlier_share_iqr(series, k=1.5):
    s = pd.to_numeric(series, errors="coerce").dropna()
    if len(s) == 0:
        return np.nan
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    mask = (s < q1 - k * iqr) | (s > q3 + k * iqr)
    return float(mask.mean()) * 100


desc = basic_summary(df_focus, num_cols)

neg_cols = ["gdp", "population", "co2", "primary_energy_consumption"]
neg = count_negatives(df_focus, neg_cols)

outlier_table = pd.DataFrame(
    {
        "feature": num_cols,
        "outliers_z_%": [round(outlier_share_z(df_focus[c]), 2) for c in num_cols],
        "outliers_iqr_%": [round(outlier_share_iqr(df_focus[c]), 2) for c in num_cols],
        "negatives_count": [int(neg.get(c, 0)) for c in num_cols],
    }
).sort_values(["outliers_z_%", "outliers_iqr_%"], ascending=False, na_position="last")

desc, outlier_table
```

**Вывод**

Показатели CO₂ и GDP содержат умеренное количество выбросов, что отражает различия в экономических масштабах стран. Методы Z-score и IQR дают схожие результаты, подтверждая наличие единичных экстремальных значений, но без критических искажений.  
Отрицательные значения отсутствуют, данные корректны для дальнейшего анализа.


#### Распределение и выбросы по ключевым экономическим и экологическим показателям  

Визуализируем распределения CO₂, ВВП и энергопотребления по группам стран,  
чтобы выявить различия в масштабах и наличие экстремальных значений.


```python
import plotly.express as px
import pandas as pd


def plot_box_log(df, cols):
    melted = df.melt(
        id_vars=["Name", "group"],
        value_vars=cols,
        var_name="feature",
        value_name="value",
    )

    fig = px.box(
        melted,
        x="group",
        y="value",
        color="group",
        facet_col="feature",
        facet_col_wrap=2,
        points="outliers",
        log_y=True,
        title="Распределения и выбросы по ключевым признакам (лог-шкала по Y)",
        category_orders={"group": ["Развивающиеся", "Переходные", "Развитые"]},
    )

    fig.update_layout(
        height=600,
        title_x=0.02,
        margin=dict(l=60, r=40, t=70, b=60),
        font=dict(size=13),
    )

    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    fig.update_xaxes(tickangle=25)
    fig.show()


num_cols = [
    c
    for c in ["co2", "co2_per_capita", "gdp", "primary_energy_consumption"]
    if c in df_focus.columns
]

plot_box_log(df_focus, num_cols)
```

График демонстрирует распределение ключевых показателей (CO₂, CO₂ per capita, GDP и энергопотребления) по трем группам стран с использованием логарифмической шкалы.

4. **Энергопотребление (primary energy consumption)**
Данный показатель коррелирует с ВВП и выбросами CO₂: чем выше уровень экономики,
тем выше потребление энергии и соответствующие выбросы.

Распределения подтверждают гипотезу о различиях в динамике развития. Развитые страны демонстрируют стабилизацию выбросов при росте ВВП, тогда как переходные продолжают фазу индустриализации с повышенной углеродоёмкостью.

## 6. Проверка взаимосвязей и согласованности данных

На этом этапе проводится проверка взаимосвязей и согласованности данных. Оценивается, как экономические, демографические и экологические показатели (ВВП, выбросы CO₂, энергопотребление и население) связаны между собой. Корреляция показывает, насколько изменения одного показателя сопровождаются изменением другого:
1. положительная корреляция означает, что значения растут вместе;
2. отрицательная — что при росте одного показатель другого уменьшается.

Такой анализ помогает количественно оценить силу связи и убедиться, что данные логичны, согласованы между собой и отражают реальные зависимости между экономическим ростом и выбросами CO₂.

### Инструменты, используемые на этом шаге:

- df.corr() — расчёт коэффициентов корреляции Пирсона;
- sns.heatmap() — визуализация корреляций в виде тепловой карты;
- msno.heatmap() — (опционально) анализ «корреляции пропусков»;
при необходимости — sns.pairplot() для парных сравнений признаков.

#### Тепловая карта корреляций

Для количественной оценки взаимосвязей между ключевыми экономическими и экологическими показателями построена тепловая карта коэффициентов корреляции Пирсона.  



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


def plot_correlation_heatmap(df, cols):
    corr = df[cols].corr(numeric_only=True)
    fig, ax = plt.subplots(figsize=(7, 5))
    sns.heatmap(
        corr,
        annot=True,
        fmt=".2f",
        cmap="coolwarm",
        vmin=-1,
        vmax=1,
        cbar_kws={"shrink": 0.8},
        square=True,
        annot_kws={"size": 10},
        ax=ax,
    )
    ax.set_title("Корреляции между ключевыми показателями", fontsize=13, pad=15)
    fig.tight_layout()
    plt.show()


cols_check = [
    "gdp",
    "co2",
    "co2_per_capita",
    "primary_energy_consumption",
    "population",
]
corr_cols = [c for c in cols_check if c in df_focus.columns]

plot_correlation_heatmap(df_focus, corr_cols)
```

Тепловая карта демонстрирует сильную положительную корреляцию между показателями ВВП, выбросов CO₂ и первичного энергопотребления (r ~ 0.9–1.0), что подтверждает их тесную взаимосвязь и отражает зависимость роста экономики от энергопотребления. Корреляция с населением умеренная (r ~ 0.4–0.5), а связь с CO₂ на душу населения слабее, что указывает на различия в уровне энергоэффективности и структуре экономик разных стран.


#### Матрица парных рассеяний

Чтобы увидеть форму взаимосвязей между показателями, построим парные диаграммы рассеяния для CO₂, CO₂ per capita, GDP и энергопотребления.


```python
import plotly.express as px


def plot_spolm(df):
    dims = [
        c
        for c in ["co2", "co2_per_capita", "gdp", "primary_energy_consumption"]
        if c in df.columns
    ]
    d = df.dropna(subset=dims)

    fig = px.scatter_matrix(
        d,
        dimensions=dims,
        color="group",
        hover_name="Name",
        title="Парная матрица рассеяния по ключевым показателям (по группам)",
    )
    fig.update_traces(diagonal_visible=True, showupperhalf=False)
    fig.update_layout(height=700, margin=dict(l=40, r=20, t=70, b=40))
    fig.show()


plot_spolm(df_focus)
```

В парах (GDP, CO₂) и (энергопотребление, CO₂) видна вытянутая диагональ, связь близка к линейной, особенно у переходных экономик.  
Для (CO₂ per capita и population) облака более круглые и связь слабее, так как уровень на душу населения больше отражает энергоэффективность и структуру экономики, чем просто численность.  
Отдельные удалённые точки это потенциальные выбросы или особые периоды (кризисы, скачки добычи).  
График не противоречит тепловой карте и поддерживает гипотезу о стадиях развития групп стран.


## 7. Проверка структуры и целостности данных с помощью Pandera
Перед финальными визуализациями важно убедиться, что таблица **df_focus** соответствует ожидаемой структуре (все столбцы присутствуют, типы корректны, значения — в допустимых пределах).

## 8. Глобальные и групповые тренды выбросов CO₂

Перед анализом отдельных стран необходимо рассмотреть, как менялись суммарные выбросы CO₂ в мировом масштабе и в разрезе трёх групп экономик. 


#### 8.1 Глобальная динамика выбросов CO₂

На этом этапе визуализируется общая мировая динамика выбросов CO₂ за период 1850–2023 гг. 


```python
import plotly.express as px


def plot_global_co2_trend_interactive(df, year_col="year", co2_col="co2"):
    if co2_col in df.columns and year_col in df.columns:
        yearly = df.groupby(year_col, as_index=False)[co2_col].sum()
        fig = px.line(
            yearly,
            x=year_col,
            y=co2_col,
            title="Глобальный тренд выбросов CO₂ (1850–2023 гг.)",
            labels={year_col: "Год", co2_col: "Глобальные выбросы CO₂ (тонны)"},
        )
        fig.update_traces(mode="lines+markers", line=dict(width=2), marker=dict(size=4))
        fig.update_layout(
            height=500,
            title_x=0.02,
            margin=dict(l=60, r=40, t=60, b=40),
            font=dict(size=13),
            hovermode="x unified",
            template="plotly_white",
        )
        fig.show()
    else:
        print(f"Колонки '{year_col}' или '{co2_col}' не найдены в датафрейме")


# вызов
plot_global_co2_trend_interactive(df_focus)
```

Глобальная кривая показывает устойчивый рост выбросов CO₂ на протяжении последних 170 лет.
Особенно резкое увеличение наблюдается во второй половине XX века, что может быть связано с индустриализацией, ростом населения и потреблением энергии. После 2000-х годов темпы роста начинают замедляться, отражая повышение энергоэффективности и развитие низкоуглеродных технологий.


#### 8.2 Динамика выбросов CO₂ по группам стран

Сравниваются три категории экономик (развитые, переходные и развивающиеся) по совокупным объёмам выбросов для того, чтобы понять какие группы вносят наибольший вклад в глобальные эмиссии.

```python
def plot_group_co2_trends(df, year_col="year", co2_col="co2"):
    if all(col in df.columns for col in [year_col, co2_col, "group"]):
        yearly = (
            df.groupby([year_col, "group"], as_index=False)[co2_col]
            .sum()
            .rename(columns={co2_col: "total_co2"})
        )

        fig = px.line(
            yearly,
            x=year_col,
            y="total_co2",
            color="group",
            title="Динамика совокупных выбросов CO₂ по группам стран (1850–2023 гг.)",
            labels={year_col: "Год", "total_co2": "Суммарные выбросы CO₂ (тонны)"},
        )

        fig.update_traces(mode="lines+markers", line=dict(width=2))
        fig.update_layout(
            height=500,
            title_x=0.02,
            margin=dict(l=60, r=40, t=60, b=40),
            font=dict(size=13),
            hovermode="x unified",
            template="plotly_white",
        )
        fig.show()
    else:
        print("Не найдены нужные колонки в датафрейме")


# вызов
plot_group_co2_trends(df_focus)
```

Развитые страны вносят наибольший вклад в глобальные выбросы, однако начиная с 1990-х годов их темпы роста стабилизируются.
Переходные экономики, прежде всего Китай и Россия, демонстрируют ускоренное увеличение выбросов до 2010-х годов.
Развивающиеся страны пока вносят меньший вклад, но темпы роста их эмиссий ускоряются по мере экономического роста.
Такая динамика отражает различия в стадиях индустриализации и уровне энергоэффективности.


#### 8.3 Динамика выбросов CO₂ по отдельным странам

Анализируются траектории выбросов CO₂ для каждой страны в составе своей экономической группы, чтобы показать различия в динамике развития.

```python
import plotly.express as px


def plot_country_co2_trends(df, year_col="year", co2_col="co2"):
    cols = [year_col, co2_col, "Name", "group"]
    data = df[[c for c in cols if c in df.columns]].dropna(subset=[year_col, co2_col])
    data = data[data[year_col].between(1960, 2023)]

    fig = px.line(
        data,
        x=year_col,
        y=co2_col,
        color="Name",
        facet_col="group",
        facet_col_wrap=3,
        title="Динамика выбросов CO₂ по странам (1960–2023 гг.)",
        labels={
            year_col: "Год",
            co2_col: "Выбросы CO₂ (тонны)",
            "Name": "Страна",
            "group": "Группа стран",
        },
        template="plotly_white",
    )

    fig.update_traces(mode="lines", line=dict(width=2))
    fig.update_layout(
        height=550,
        title_x=0.02,
        margin=dict(l=60, r=40, t=60, b=40),
        font=dict(size=13),
        hovermode="x unified",
        legend_title_text="Страны",
    )
    fig.update_xaxes(title="Год")
    fig.update_yaxes(title="Выбросы CO₂ (тонны)")
    fig.show()


plot_country_co2_trends(df_focus)
```

У развитых стран отмечается стабилизация и постепенное снижение эмиссий при продолжающемся росте ВВП.
В переходных экономиках сохраняется линейная зависимость между экономическим ростом и выбросами, при этом наблюдаются отдельные периоды ускорения, связанные с индустриализацией.
У развивающихся стран уровень выбросов остаётся низким, однако темпы их увеличения растут по мере расширения производства и энергопотребления.


**Общий вывод** 

Анализ временных рядов показал, что динамика выбросов CO₂ на глобальном уровне и по группам стран подчиняется схожей закономерности.
Мировые выбросы росли на протяжении XIX–XX веков, достигая максимума к началу XXI века, после чего темпы роста начали снижаться.
Развитые экономики демонстрируют признаки стабилизации выбросов при росте ВВП, что указывает на повышение энергоэффективности.
Переходные страны продолжают путь интенсивного индустриального роста, а развивающиеся экономики находятся на этапе расширения, сопровождающегося увеличением выбросов.
Совокупная картина подтверждает существование различных стадий энергетического и технологического развития, определяющих вклад стран в глобальные эмиссии.

### 9. Анимационная карта динамики выбросов CO₂ по странам

Для пространственно-временного анализа используем интерактивную карту, показывающую изменение выбросов CO₂ в мире с 1960 по 2023 год. Размер точки отражает уровень выбросов, цвет — принадлежность к стране.



```python
import plotly.express as px
import pycountry


def add_iso3(df):
    df["iso_code"] = df["Name"].map(
        lambda x: (
            pycountry.countries.get(name=x).alpha_3
            if pycountry.countries.get(name=x)
            else None
        )
    )
    return df


if "iso_code" not in df_focus.columns:
    df_focus = add_iso3(df_focus)


def plot_geo_co2(df, value_col="co2"):
    df_map = df.dropna(subset=["year", value_col, "iso_code"])
    df_map = df_map[df_map["year"].between(1960, 2023)]
    fig = px.scatter_geo(
        df_map,
        locations="iso_code",
        locationmode="ISO-3",
        color="Name",
        hover_name="Name",
        size=value_col,
        animation_frame="year",
        projection="natural earth",
        title="Динамика выбросов CO₂ по странам мира (1960–2023)",
    )
    fig.update_layout(
        height=520,
        margin=dict(l=40, r=40, t=60, b=40),
        font=dict(size=13),
        legend_title_text="Страна",
    )
    fig.show()


plot_geo_co2(df_focus)
```

 
На карте хорошо видно, как в конце XX века центр мировых выбросов 
постепенно смещается из Европы и США в Азию.  
Китай и Индия демонстрируют стремительный рост выбросов, 
тогда как развитые страны постепенно стабилизируют или снижают уровень CO₂.


## 10. Зависимость выбросов CO₂ на душу населения от ВВП
Построение анимированной диаграммы рассеяния, показывающую взаимосвязь между уровнем ВВП и выбросами CO₂ на душу населения во времени.
Размер точек отражает численность населения, цвет — принадлежность к стране.

```python
import plotly.express as px


def plot_gdp_vs_co2pc_animated(df, year_range=(1960, 2023)):
    cols = ["year", "Name", "group", "gdp", "co2_per_capita", "population"]
    d = df[[c for c in cols if c in df.columns]].dropna(
        subset=["gdp", "co2_per_capita"]
    )

    y0, y1 = year_range
    d = d[(d["year"] >= y0) & (d["year"] <= y1)]

    fig = px.scatter(
        d,
        x="gdp",
        y="co2_per_capita",
        color="Name",
        hover_name="Name",
        size="population",
        animation_frame="year",
        log_x=True,
        opacity=0.75,
        size_max=55,
        title=f"CO₂ на душу населения vs ВВП ({y0}–{y1}) по странам",
        labels={"gdp": "ВВП (лог)", "co2_per_capita": "CO₂ на душу населения"},
    )

    fig.update_layout(
        title_x=0.02,
        margin=dict(l=50, r=40, t=70, b=60),
        font=dict(size=13),
        height=550,
    )
    fig.show()


plot_gdp_vs_co2pc_animated(df_focus)
```


Видно, что между показателями ВВП и выбросами CO₂ на душу населения сохраняется устойчивая положительная зависимость.
У развитых стран (США, Германия, Япония) рост ВВП постепенно перестаёт сопровождаться увеличением выбросов — кривая выравнивается, что указывает на начало «отвязки» экономического роста от эмиссий.
В переходных экономиках (Китай, Россия, Индия) связь остаётся почти линейной: по мере роста ВВП выбросы также увеличиваются.
Развивающиеся страны (Нигерия, Индонезия, Пакистан) движутся вдоль нижней части графика, демонстрируя низкий ВВП и умеренные, но растущие выбросы.
Такое распределение подтверждает гипотезу о разных стадиях экологического и экономического развития стран.

## 11. Связь энергопотребления и выбросов парниковых газов (GHG)
Необходимо проверить, существует ли прямая зависимость между уровнем энергопотребления и общими выбросами парниковых газов по странам.
Это помогает понять, насколько экономическое развитие и энергетическая активность связаны с экологической нагрузкой.

#### Инструменты, использованные на этом шаге:

1. plotly.express.scatter() — визуализация зависимости между энергопотреблением и выбросами;
2. Логарифмическая шкала по оси X — для выравнивания диапазонов значений;
3. Группировка по категориям стран для сопоставления тенденций.

Сравним страны в один выбранный год, чтобы увидеть форму зависимости между
энергопотреблением и совокупными выбросами парниковых газов (GHG).

```python
import plotly.express as px


def plot_energy_vs_ghg(df, year=2020):
    cols = ["Name", "group", "year", "primary_energy_consumption", "total_ghg"]
    data = df[[c for c in cols if c in df.columns]].dropna()
    data = data[data["year"] == year]

    fig = px.scatter(
        data,
        x="primary_energy_consumption",
        y="total_ghg",
        color="group",
        hover_name="Name",
        size="primary_energy_consumption",
        trendline="ols",
        log_x=True,
        title=f"Связь энергопотребления и выбросов GHG ({year})",
        labels={
            "primary_energy_consumption": "Энергопотребление (лог-шкала)",
            "total_ghg": "Общие выбросы парниковых газов",
        },
        category_orders={"group": ["Развивающиеся", "Переходные", "Развитые"]},
    )

    fig.update_layout(
        height=500,
        title_x=0.02,
        margin=dict(l=50, r=40, t=70, b=50),
        font=dict(size=13),
    )
    fig.show()


plot_energy_vs_ghg(df_focus, year=2020)
```

Теперь посмотрим, как страны движутся во времени в пространстве (Еnergy в GHG). Траектории отражают, растут ли выбросы вместе с потреблением,
или появляется замедление за счёт улучшения энергоэффективности.


```python
import plotly.express as px


def animate_energy_vs_ghg(df, year_range=(1960, 2023)):
    cols = ["Name", "group", "year", "primary_energy_consumption", "total_ghg"]
    d = df[[c for c in cols if c in df.columns]].dropna()
    y0, y1 = year_range
    d = d[(d["year"] >= y0) & (d["year"] <= y1)]

    fig = px.scatter(
        d,
        x="primary_energy_consumption",
        y="total_ghg",
        color="Name",
        hover_name="Name",
        size="primary_energy_consumption",
        animation_frame="year",
        log_x=True,
        size_max=50,
        opacity=0.8,
        title=f"Энергопотребление vs GHG по странам, {y0}–{y1}",
    )
    fig.update_layout(
        height=540,
        title_x=0.02,
        margin=dict(l=50, r=40, t=70, b=50),
        font=dict(size=13),
    )
    fig.show()


animate_energy_vs_ghg(df_focus)
```

Наблюдается ярко выраженная положительная зависимость:
страны с высоким энергопотреблением демонстрируют и более высокие выбросы парниковых газов.
Однако среди развитых стран (Япония, Германия, США) темпы роста выбросов ниже при сопоставимом уровне энергопотребления,
что может свидетельствовать об увеличении энергоэффективности и частичной “отвязке” экономического роста от загрязнения.

**Вывод**

На графиках прослеживается выраженная положительная зависимость между энергопотреблением и выбросами парниковых газов (GHG).
Развитые страны имеют высокий уровень энергопотребления и значительные выбросы,однако у некоторых из них (США, Германия, Япония) наблюдается замедление роста выбросов при сохранении высоких уровней потребления энергии. Это указывает на повышение энергоэффективности и частичную «отвязку» экономического роста от загрязнения атмосферы.
Переходные и развивающиеся экономики (например, Китай, Индия, Индонезия) демонстрируют более прямолинейную зависимость, что отражает сохраняющуюся тесную связь между энергетическим развитием и уровнем эмиссий.
Таким образом, структура энергобаланса и эффективность использования ресурсов остаются ключевыми факторами в процессе декарбонизации.

## 12. Сравнение средних показателей между группами стран
Для итогового сравнения агрегируем данные по трем категориям стран и визуализируем средние значения ключевых экономико-экологических показателей.
Это позволяет подтвердить различия в уровне выбросов, ВВП и энергопотребления между развитыми, переходными и развивающимися экономиками.

```python
import plotly.express as px
import pandas as pd


def plot_group_comparison(df):
    cols = ["group", "co2", "co2_per_capita", "gdp", "primary_energy_consumption"]
    d = df[[c for c in cols if c in df.columns]].dropna()
    agg = d.groupby("group").mean(numeric_only=True).reset_index()

    melted = agg.melt(
        id_vars="group",
        value_vars=["co2", "co2_per_capita", "gdp", "primary_energy_consumption"],
        var_name="Показатель",
        value_name="Среднее значение",
    )

    fig = px.bar(
        melted,
        x="Показатель",
        y="Среднее значение",
        color="group",
        barmode="group",
        title="Сравнение групп стран по ключевым показателям",
        category_orders={"group": ["Развивающиеся", "Переходные", "Развитые"]},
    )
    fig.update_layout(
        height=500,
        title_x=0.02,
        margin=dict(l=50, r=40, t=70, b=60),
        font=dict(size=13),
    )
    fig.update_xaxes(tickangle=25)
    fig.show()


plot_group_comparison(df_focus)
```

**Вывод**

Средние значения подтверждают различие между группами:
Развитые страны — лидируют по ВВП, но их выбросы CO₂ на душу населения стабилизируются.
Переходные экономики (Китай, Индия, Россия) показывают рост по всем параметрам.
Развивающиеся страны остаются на низких уровнях ВВП и энергопотребления,
но темпы роста выбросов относительно их масштабов остаются высокими.
Эта разница отражает разный этап энергетического и экономического развития,
а также степень внедрения технологий энергоэффективности и декарбонизации.
